```python
import pandas as pd
import json
import matplotlib.pyplot as plt
plt.style.use('default')
```

```python
df = pd.read_csv('../data/CarvanaClean.csv')
```

```python
df.head()
```

```python
df.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
df.head(2)
```

```python
categorical_cols = list(df[['model', 'details', 'make']].columns)
```

```python
categorical_cols
```

```python
def encode_categories(df, columns):
    for col in columns:
        df[f'{col}_c'] = pd.Categorical(df[col])
        df[f'{col}_cat'] = df[f'{col}_c'].cat.codes
        mapping = dict( enumerate(df[f'{col}_c'].cat.categories))
        with open(f'./list-data/{col}_listData.json', 'w') as file:
            json.dump(mapping, file)
        df.drop(f'{col}_c', axis=1, inplace=True)
        df.drop(col, axis=1, inplace=True)

    return df
```

```python
encoded_df = encode_categories(df, categorical_cols)
```

```python
encoded_df
```

```python
encoded_df.to_csv('../data/CarvanaEncoded.csv')
```
